# Performance analysis

## Modules

```python
import pandas as pd
from tabulate import tabulate
```

```python

# Function to read a CSV file into a pandas DataFrame
def read_csv_to_dataframe(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None

# Function to print DataFrame using tabulate
def print_dataframe(df):
    if df is not None:
        print(tabulate(df, headers='keys', tablefmt='psql'))
    else:
        print("No data to display.")
```

```python
# Paths to the CSV files
single_image_csv_path = 'single_image_performances.csv'
dataset_eval_csv_path = 'dataset_eval_performances.csv'

# Read the CSV files into DataFrames
single_image_df = read_csv_to_dataframe(single_image_csv_path)
dataset_eval_df = read_csv_to_dataframe(dataset_eval_csv_path)

# Print the DataFrames
print("Single Image Performances:")
print_dataframe(single_image_df)

print("\nDataset Evaluation Performances:")
print_dataframe(dataset_eval_df)
```
